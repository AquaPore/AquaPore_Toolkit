```julia
Ks_New(σ) = (1. /  (100.0 .* exp.(σ .* 3.0))) * exp.((σ .^ 2.0) ./ 2.0)
```

```julia

σ = fill(0.0::Float64, 100)
Ks = fill(0.0::Float64, 100)
 σ[1] = 0.8
for i=2:100
   σ[i] =  σ[i-1] + (4.0 - 0.8) / 100.0
   Ks[i] = Ks_New(σ[i])
end
```

```julia
f = Figure()
ax = Axis(f[1, 1])

lines!(ax, σ, Ks)
f

   # scatter!(Fig, σ, Ks)
```
