```julia
		T2_Max = 3.0
		   T2_Min = 1.0
			T2(τ₂) = (T2_Min - T2_Max) * τ₂ + T2_Max
			T3(P) = (T2_Max-T2_Min ) * P + T2_Min
Ks_New(σ, τ₂, P) = (0.5*((1. /  (100.0 .* exp.(σ .* 3.0))) .^ T2(τ₂)) * exp.(((σ .* T2(τ₂) ) .^ 2.0) ./ 2.0))^T3(P)
```

```julia

σ = fill(0.0::Float64, 100)
Ks_1 = fill(0.0::Float64, 100)
Ks_2 = fill(0.0::Float64, 100)
 σ[1] = 0.8
for i=2:100
   σ[i] =  σ[i-1] + (4.0 - 0.8) / 100.0
   Ks_1[i] = Ks_New(σ[i],0.5, 1.0)
   Ks_2[i] = Ks_New(σ[i],0.5, 1.2)
end
```

```julia
f = Figure()
ax = Axis(f[1, 1])


lines!(ax, σ, Ks_1, color = :blue)
lines!(ax, σ, Ks_2, color = :red)
f

   # scatter!(Fig, σ, Ks)
```
